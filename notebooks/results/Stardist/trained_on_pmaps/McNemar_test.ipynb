```python
import os
import pandas as pd
import numpy as np
from tifffile import imread
from tqdm import tqdm

import sys
from pathlib import Path
from mlxtend.evaluate import mcnemar_table
from mlxtend.evaluate import mcnemar

cwd = os.getcwd()
# print(f"Working dir: {cwd}")
module_path = os.path.abspath(Path(cwd).parents[3])
# print(f"Module dir: {module_path}")
if module_path not in sys.path:
    sys.path.append(module_path)

import seaborn as sns
sns.set_theme(style="darkgrid")
sns.set_palette("Set2")
```

```python
df_4 = pd.read_csv(os.path.join(module_path, "experiments", "exp12-4", "results-defaults", "288x288", "reg_props" ,"all.csv"))
```

```python
df_4 = df_4.set_index("label", drop=False)
df_4n = df_4[["label", "actual_frame","IoU"]]
clm=["lbl", "actual_frame","IoU"]
df_4n.columns=clm
df_4n
```

```python
df_11 = pd.read_csv(os.path.join(module_path, "experiments", "exp11", "results_defaults", "reg_props", "all.csv"))
```

```python
df_11 = df_11.set_index("label", drop=False)
df_11n = df_11[["label", "actual_frame","IoU"]]
clm=["lbl", "actual_frame","IoU"]
df_11n.columns=clm
df_11n
```

```python
result=pd.merge(df_11n,df_4n,how="left", on=["lbl", "actual_frame"], indicator=True)
#result=df_4n.compare(df_11n)
pd.set_option('display.max_rows', result.shape[0]+1)
result
```

```python
arr_4= np.where(result["IoU_y"].isna(), 0, 1)
arr_4
```

```python
arr_11= np.where(result["IoU_x"].isna(), 0, 1)
arr_11
```

```python
arr_gt = np.where(result["IoU_x"]>0.3, 1, 0)
arr_gt
```

```python
tb = mcnemar_table(y_target=arr_gt, 
                   y_model1=arr_4, 
                   y_model2=arr_11)
tb
```

```python
chi2, p = mcnemar(ary=tb, corrected=True)
print('chi-squared:', chi2)
print(f'p-value: {p:.8f}')
```

```python
chi2, p = mcnemar(ary=tb, corrected=False)
print('chi-squared:', chi2)
print(f'p-value: {p:.8f}')
```

```python
# interpret the p-value
alpha = 0.05
if p > alpha:
	print('Same proportions of errors (fail to reject H0)')
else:
	print('Different proportions of errors (reject H0)')
```
